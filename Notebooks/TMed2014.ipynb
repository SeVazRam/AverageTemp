```python
data_path = os.getcwd()+ '/Data/Temp/2014/'
db_data_path = data_path 
os.listdir()
print(os.listdir(db_data_path))
```

```python
path = os.getcwd()+ '/Data/Temp/2014/'
all_files = glob.glob(path+'*.csv')
all_files
```

```python
#dfPrueba=[df1,df2]
li = []
TMed = ['TMed201401','TMed201402','TMed201403','TMed201404','TMed201405','TMed201406','TMed201407','TMed201408','TMed201409',
        'TMed201410','TMed201411','TMed201412' ]
```

```python
for i in range(len(TMed)):
    df = pd.read_csv(all_files[i], encoding='latin-1')
    df = df.rename(columns={'Tmed':TMed[i]})
    li.append(df)
    
li[0].head()
```

```python
d1=li[0]
d2=li[1]
d3=li[2]
d4=li[3]
d5=li[4]
d6=li[5]
d7=li[6]
d8=li[7]
d9=li[8]
d10=li[9]
d11=li[10]
d12=li[11]
#len(d1),len(d2)
```

```python
#d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12
```

```python
d12
```

```python
df_outer = pd.merge(d1,d2,
              on=['Clave','Lon','Lat','Est','Edo'], how='outer')

df_outer = pd.merge(df_outer , d3, on=['Clave','Lon','Lat','Est','Edo'], how='outer')

#pd.concat(li,axis=0,ignore_index=True)
df_outer.head()
```

```python
from functools import reduce
dfl=[d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Clave','Lon','Lat','Est','Edo'],
                                            how='outer'), dfl)

df_merged
```

```python
#df_merged = df_merged.drop(2034)
```

```python
df_merged.info()
```

```python
cols = ["TMed201401","TMed201402","TMed201403","TMed201404","TMed201405","TMed201406","TMed201407","TMed201408",
       "TMed201409","TMed201410","TMed201411","TMed201412"]
df_merged['TMed'] = df_merged[cols].mean(axis=1)
df_merged
```

```python
geo_df1 = gpd.GeoDataFrame(df_merged, geometry=gpd.points_from_xy(df_merged.Lon, df_merged.Lat), crs="EPSG:4326")
geo_df1
```

```python
intersect = gpd.overlay(geo_df1, geo_df, how='intersection')
intersect
```

```python
grouped = intersect.groupby(['id','state'])['TMed'].agg('mean')
grouped.head()
```

```python
merge = pd.merge(geo_df, grouped, on="id")
merge = merge.rename(columns={'TMed': 'TMed14'})
merge.head()
```

```python
merge['TMed14'].max()
```

```python
import matplotlib.pyplot as plt

merge.plot(column='TMed14', figsize=(10,10),
           legend_kwds={'label': "TMed2014 per State", 'orientation': "horizontal"}, legend=True, cmap='OrRd',vmin=14.5, vmax=29.2)

plt.grid(linestyle = '--', linewidth = 0.5)
plt.xlabel("Longitud")
plt.ylabel("latitude")
plt.title("Temperature by state")
plt.savefig("TMed2014")
```
